```
students = KTU_rdd.map(lambda x : x.strip()).map(lambda x : ' '.join(x.split())).filter(lambda x : len(x.split(' ')) >=10).filter(lambda x : "Register" not in x).map(lambda x : x.split(' '))
```

```
#Total Number of Students 

students.count()
```

```
students_df = sqlContext.createDataFrame(students,['id','sub1','sub2','sub3','sub4','sub5','sub6','sub7','sub8','sub9'])
```

```
students_df.take(10)
```

```
students_df.registerTempTable("results")
```
