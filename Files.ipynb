```python
#Word count of the book Alice in wonderland

f = open("C:\Users\kskrishnasangeeth\Desktop\chennaipy\\alice.txt",'r')
words = f.read().split()
```

```python
wordDict={}
for each_word in words:
    if not each_word in wordDict:
        wordDict[each_word]=1
    else:
        wordDict[each_word]+=1

for word,count in wordDict.items():
    print word,'->',count
```
