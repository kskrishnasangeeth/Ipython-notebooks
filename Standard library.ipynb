```python
#Serializing python objects using pickle

import pickle
```

```python
numList = [ num for num in range(1,100)]
print numList
```

```python
#writing the serialized python objects to list.pi

pickle.dump(numList,open("list.pi","wb"))
```

```python
#Deserializing the python objects

pickle.load(open("list.pi","rb"))
```

```python
#Return the pickled representation of the object as a string, instead of writing it to a file.
myTuple=('Krishna','Sangeeth')
obj = pickle.dumps(myTuple)

print pickle.loads(obj)

```

```python
#Let's start with a simple DB

import sqlite3

conn = sqlite3.connect('test.db')


print "Opened database successfully";

conn.execute('''CREATE TABLE COMPANY
       (ID INT      NOT NULL,
       NAME           TEXT    NOT NULL,
       AGE            INT     NOT NULL,
       ADDRESS        CHAR(50),
       SALARY         REAL);''')
print "Table created successfully";

conn.close()
```

```python
#inserting values into table 

import sqlite3

conn = sqlite3.connect('test.db')
print "Opened database successfully";

conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY) \
      VALUES (1, 'Paul', 32, 'California', 20000.00 )");

conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY) \
      VALUES (2, 'Allen', 25, 'Texas', 15000.00 )");

conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY) \
      VALUES (3, 'Teddy', 23, 'Norway', 20000.00 )");

conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY) \
      VALUES (4, 'Mark', 25, 'Rich-Mond ', 65000.00 )");

conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY)  VALUES (?,?,?,?,?)", (5,'John',30,'Virginia',9000.00))

conn.commit()
print "Records created successfully";
conn.close()
```

```python
#retrieving values from the table 

import sqlite3

conn = sqlite3.connect('test.db')
print "Opened database successfully";

cursor = conn.execute("SELECT id, name, address, salary  from COMPANY")
for row in cursor:
   print "ID = ", row[0]
   print "NAME = ", row[1]
   print "ADDRESS = ", row[2]
   print "SALARY = ", row[3], "\n"

print "Operation done successfully";
conn.close()
```
